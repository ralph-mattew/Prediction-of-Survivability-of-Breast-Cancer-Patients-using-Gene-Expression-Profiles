```python
raw['death_from_cancer'].fillna('Living', inplace=True)
df = raw[raw['death_from_cancer'] != 'Died of Other Causes'].iloc[:, :515].copy()
df.drop(columns=['death_from_cancer'], inplace=True)
```

```python

def discretize(x):
    y = x/12
    if y <= 5:
        return 0
    else:
        return 1


df['time_bin'] = df['overall_survival_months'].apply(lambda x: discretize(x))
df.groupby('time_bin')['overall_survival'].value_counts().plot(kind='bar')
```

```python
# remove irrelevant and redundant features
df.drop(columns=['patient_id', 'cancer_type', 'er_status_measured_by_ihc', 'her2_status'], inplace=True)
```

```python
# map cellularity
cellularity_map = {'High': 3, 'Moderate': 2, 'Low': 1}
df['cellularity'] = df['cellularity'].map(cellularity_map)
```

```python
df['neoplasm_histologic_grade'].fillna(df['neoplasm_histologic_grade'].value_counts().index[0], inplace=True)
df['mutation_count'].fillna(df['mutation_count'].value_counts().index[0], inplace=True)
df['tumor_size'].fillna(df['tumor_size'].value_counts().index[0], inplace=True)
df['tumor_stage'].fillna(df['tumor_stage'].value_counts().index[0], inplace=True)
df['cellularity'].fillna(df['cellularity'].value_counts().index[0], inplace=True)
```

```python
df.isna().sum() [df.isna().sum() >0 ]
```

```python
df['type_of_breast_surgery'].fillna('others', inplace=True)
df['cancer_type_detailed'].fillna('others', inplace=True)
df['tumor_other_histologic_subtype'].fillna('others', inplace=True)
df['primary_tumor_laterality'].fillna('others', inplace=True)
df['oncotree_code'].fillna('others', inplace=True)
df['3-gene_classifier_subtype'].fillna('others', inplace=True)
```

```python
df['tumor_size'].plot(kind='density')
```

```python
df['log_tumor_size'] = (df['tumor_size']+1).transform(np.log)

df.drop(columns=['tumor_size'], inplace=True)
```

```python
df = df.replace('Breast', df['cancer_type_detailed'].value_counts().index[0])
```

```python
df = df[df['time_bin'] == 0].copy()
```

```python
from sklearn.svm import SVR
```

```python
## one-hot encode nominal features 
X = pd.get_dummies(df.drop(columns=['overall_survival', 'overall_survival_months']))
y = df['overall_survival_months']
# 'log_tumor_size', 'nottingham_prognostic_index'
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
lr = SVR(kernel='rbf', C=500, gamma='auto', epsilon=1).fit(X_train, y_train)
print("training set score: %f" % lr.score(X_train, y_train))
print("test set score: %f" % lr.score(X_test, y_test))
```

```python

p = 0.9
sel = VarianceThreshold(threshold=(p * (1 - p)))
X_reduced_variance = sel.fit_transform(X.values)
print('Size before Variance Threshold reduction: ', X.values.shape)
print(f'Estimated size after {p*100}% Variance Threshold reduction: ', 
      X_reduced_variance.shape)
```

```python
svd = TruncatedSVD(n_components=524).fit(X)
X_reduced = svd.transform(X)
explained_var = svd.explained_variance_
explained_var_ratio = svd.explained_variance_ratio_
s_values = svd.singular_values_
```

```python
fig, (ax1, ax2) = plt.subplots(1,2,figsize=(12,5))

ax1.plot(range(1, len(explained_var_ratio)+1), explained_var_ratio, 'o-')
# ax1.set_ylim(0,1)
ax1.set_xlabel('SVD')
ax1.set_ylabel('variance explained')
ax1.set_title('Variance Explained vs No. of Components');

ax2.plot(range(1, len(explained_var_ratio)+1), explained_var_ratio.cumsum(), 
         'o-')
# ax2.set_ylim(0,1)
ax2.set_xlabel('number of Principal Components')
ax2.set_ylabel('cumulative variance explained')
ax2.set_title('Cumulative Variance Explained vs No. of Components')
ax2.axhline(y=0.9)
ax2.axvline(x=210)
plt.tight_layout()
```

```python
svd = TruncatedSVD(n_components=210).fit(X)
X_reduced = svd.transform(X)
explained_var = svd.explained_variance_
explained_var_ratio = svd.explained_variance_ratio_
s_values = svd.singular_values_
```

```python
X_reduced = StandardScaler().fit_transform(X_reduced)
```

```python
i = np.identity(X.shape[1])
weights = pd.DataFrame(svd.transform(i), index=X.columns)
weights.columns = [f'PC-{i+1}' for i in weights.columns]
weights = weights.transform(np.abs).copy()

fig, ax = plt.subplots(1,1, figsize=(8,8))
sns.heatmap(weights.sort_values(by='PC-1', ascending=False).iloc[:30, :15], 
            ax=ax)
ax.set_title('Feature Weights in Top 15 Principal Components')
```
